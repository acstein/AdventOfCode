```python
class Board:
    def __init__(self, board):
        self.board = board
        self.mask = np.zeros_like(board)
        self.complete = False
        
    def number_called(self, number):
        match = np.where(self.board == number)
        if match:
            self.mask[match] = 1
        #Pdb().set_trace()
        #self.mask[match] = 1
        
    def check_for_win(self):
        for x in self.mask:
            if all(i == 1 for i in x):
                self.complete = True
        for x in self.mask.T:
            if all(i == 1 for i in x):
                self.complete = True
        return self.complete
    
    def result(self, last_called):
        match = np.where(self.mask == 0)
        final = self.board[match]
        
        sum_value = np.sum(final)
        return sum_value * last_called
```

```python
def split_line(line):
    return [int(num) for num in line.split() if num]
```

```python
winner = False
boards = [Board(b) for b in boards]
for n in numbers:
    for b in boards:
        if not b.complete:
            b.number_called(n)
            if b.check_for_win():
                print(b.result(n))
```
